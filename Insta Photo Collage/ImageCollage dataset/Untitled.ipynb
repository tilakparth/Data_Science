```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cv2
```

```python
img1 =cv2.imread('top_left.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img1 = cv2.resize(img1,(200,200))
img2 =cv2.imread('top_right.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img2 = cv2.resize(img2,(200,200))

img3 =cv2.imread('bottom_left.jpg')
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
img3 = cv2.resize(img3,(200,200))
img4 =cv2.imread('bottom_right.jpg')
img4 = cv2.cvtColor(img4,4)
img4 = cv2.resize(img4,(200,200))

img5 =cv2.imread('center.jpeg')
img5 = cv2.cvtColor(img5,4)
img5 = cv2.resize(img5,(100,100))
```

```python
result = np.zeros((430,430,3),dtype='int')
# print(img1)
result[10:210,10:210] = img1
result[220:420,10:210]=img3
result[10:210,220:420] = img2
result[220:420,220:420] = img4
result[155:275,155:275] = 0
result[165:265,165:265] = img5
# result[]
plt.imshow(result)
plt.show();
```
