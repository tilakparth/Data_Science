```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = x**2 + 3
```

```python
plt.style.use('dark_background')
```

```python
plt.plot(x,y,label="Stock Price")
plt.plot(x,x**1.5,label="Stock Pricejnnkbk")
plt.xlabel("Timestamp");
plt.title("My line plot")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
plt.scatter(x,y,label="Stock Price")
plt.scatter(x,x**1.5,label="Stock Pricejnnkbk")
plt.xlabel("Timestamp");
plt.title("My line plot")
plt.ylabel("price")
plt.legend()
plt.show()
```
