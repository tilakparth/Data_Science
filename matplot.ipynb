```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y = x**2 - 5*x
```

```python
print(plt.style.available)
plt.style.use('dark_background')
```

```python
plt.plot(x,y,label="Stock Price",color="red")
# plt.show()
plt.plot(x,x*1.5,"g*--",label="Stock Pricejnnkbk")
plt.xlabel("Timestamp");
plt.title("My line plot")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
plt.scatter(x,y,label="Stock Price",marker="^")
plt.scatter(x,x**1.5,color="green",label="Stock Pricejnnkbk",marker="*")
plt.xlabel("Timestamp");
plt.title("My line plot")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
x = np.arange(10000)
y =[]
for i in x:
    y.append(np.sin(i/1000))
    
# y = np.sin(x/1000000)
```

```python
plt.plot(x,y,"r--")
```

```python
plt.bar(x,x*1.5,label="Stock Pricejnnkbk",width=0.5,color="yellow")
plt.bar(x+.5,abs(y)+1,label="Stock Price",color="red",width=.5)
# plt.show()

plt.xlabel("Timestamp");
plt.title("My line plot")
plt.ylabel("price")

plt.xlim(0,10)
plt.ylim(0,45)
plt.legend()
plt.show()
```
