```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = np.arange(-1,1,0.01)
b = a
```

```python
a,b = np.meshgrid(a,b)
print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection = "3d")
axes.plot_surface(a,b,b**2+a**2,cmap ="rainbow")
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection = "3d")
axes.contour(a,b,b**2+a**2,cmap ="rainbow")
plt.show()
```

```python
X = pd.read_csv("../data/hard_work_pays_off_challenge_data/Linear_X_Train.csv").values
Y = pd.read_csv("../data/hard_work_pays_off_challenge_data/Linear_Y_Train.csv").values

theta = np.load("thetalist.npy")
```

```python
plt.ion()
for i in range(30):
    y_hat = theta[i][0] + theta[i][1]*X
    plt.scatter(X,Y)
    plt.plot(X,y_hat,color='red')
    plt.draw()
    plt.show()
    plt.pause(1)
    plt.clf()
```
