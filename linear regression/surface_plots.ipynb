```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
a = np.arange(-1,1,0.01)
b = a
```

```python
a,b = np.meshgrid(a,b)
print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection = "3d")
axes.plot_surface(a,b,b**2+a**2,cmap ="rainbow")
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection = "3d")
axes.contour(a,b,b**2+a**2,cmap ="rainbow")
plt.show()
```
